write a python program to:
1.Generate a 1000000 random numbers
2.compute mean median mode using numpy functions
3.compare execution time of both

```python
import random 

import time

import numpy as np

from collections import Counter

from scipy import stats

# initialization of random numbers it means here we are generating 10 lakh random number

data= [random.randint(1,1000)for _ in range(100000)]
# lets create a numpy array

np_array= np.array(data)


# now lets start timer first

start_time=time.time()

mean_value = np.mean(np_array)
median_value = np.median(np_array)
# for mode value we need stats
mode_value = stats.mode(np_array, keepdims=True).mode[0]
# this print ModeResult(mode=array([557]), count=array([129]))
# but for only value we need to print like mode[0] oth index of mode is value

end_time=time.time()

total_timetaken= end_time-start_time
print(mean_value)
print(median_value)
print(mode_value)
print(total_timetaken)

```
